## <center> A Crash Course on NumPy for Images </center>

```python
from skimage import data
import matplotlib.pyplot as plt

camera = data.camera()
print(type(camera))    # The type of data holding the image
print(camera.shape)    # The dimensions of the image
print(camera.size)     # The number of pixels in the image

print(camera.min(), camera.max())
print(camera.mean())

plt.imshow(camera, cmap="gray")  # The dimensions is 512x512, there's no color.
```

### NumPy Indexing

```python
print(camera[10, 20])       # Check value of specific pixel
camera[3, 10] = 0           # Set specific pixel to black

camera[:10] = 10      # Set first ten rows to plack

# Masking (indexing with masks of booleans):
mask = camera < 87
camera[mask] = 255
plt.imshow(camera, cmap='gray')
```

```python
import numpy as np

# Indexing with sets of indices:
inds_r = np.arange(len(camera))
inds_c = 4 * inds_r % len(camera)
camera[inds_r, inds_c] = 0

print(len(camera))
print(inds_r)
print(inds_c)
```

```python
# Masking is useful to select a set of pixels on which to perform some manipulations

nrows, ncols = camera.shape
row, col = np.ogrid[:nrows, :ncols]        # Seems to create some sort of 2D array
cnt_row, cnt_col = nrows / 2, ncols / 2

outer_disk_mask = ((row - cnt_row)**2 + (col - cnt_col)**2 > (nrows / 2)**2)
camera[outer_disk_mask] = 0

plt.imshow(camera, cmap='gray')
```

```python
lower_half = row > cnt_row
lower_half_disk = np.logical_and(lower_half, outer_disk_mask)
camera = data.camera()
camera[lower_half_disk] = 0

plt.imshow(camera, cmap='gray')
```

### Color Images

```python
cat = data.chelsea()
print(type(cat))
print(cat.shape)        # Images just have an additional dimension

plt.imshow(cat)
```

```python
print(cat[10, 20])
cat[10, 20]
```

```python
# Set the pixel at [50, 60] to black:
cat[50, 60] = 0

# Set the pixel at row 50, column 61 to green
cat[50, 61] = [0, 255, 0]    # [red, green, blue]
plt.imshow(cat)
```

```python
# You can also use a 2D boolean mask for a 2D color image:
from skimage import data
cat = data.chelsea()
reddish = cat[:, :, 0] > 160      # Wherever the red channel is > 160
cat[reddish] = [0, 255, 0]        # We make it green
plt.imshow(cat)
```

### Coordinate Conventions

For color images, the coordinates start from the top left and are referenced as (row, column, channel). As opposed to (x, y, z) like in Matplotlib for Cartesian coordinates.

For 3D images, like videos or MRI scans, the leading dimension is a <i>plane</i>, abbreviated to <i>pln</i> or <i>p</i>. If the 3D image is grayslace, the dimensions are then (pln, row, col).

```python
# You can use scikit functions on 3D images
from skimage import morphology
from scipy import ndimage as ndi

im3d = np.random.rand(100, 1000, 1000)    # 3D array of shape (100, 1000, 1000)
print(im3d.shape)

seeds = ndi.label(im3d < 0.1)[0]
print(seeds)

#ws = morphology.watershed(im3d, seeds)
```

```python
# The third imaging dimension has lower resolution, usually.
# You can use the spacing keyworkd to process these images:
from skimage import segmentation

slics = segmentation.slic(im3d, spacing=[5, 1, 1], multichannel=False)

print(slics)
```

```python
# Other times, processing must be done plane-wise:
from skimage import filters
edges = np.zeros_like(im3d)
for pln, image in enumerate(im3d):
    # Iterate over the leading dimension (planes)
    edges[pln] = filters.sobel(image)
```
